# MATPLOTLIB

```python
# Importing Matplotlib's pyplot module for creating visualizations
import matplotlib.pyplot as plt
```

## LINE PLOT

- plt.plot(x, y, color="color_name", linestyle="line_style", linewidth=value, marker="marker_symbol", markersize=value, alpha=value, label="label_name") ---> USED TO CREATE A LINE PLOT.
    - alpha=value ---> SETS THE TRANSPARENCY LEVEL OF THE PIE SLICES (0 = FULLY TRANSPARENT, 1 = FULLY OPAQUE).
- plt.show() ---> USED TO DISPLAY ALL CREATED PLOTS ON THE SCREEN.
- plt.xlabel("text") ---> USED TO SET A LABEL FOR THE X-AXIS.
- plt.ylabel("text") ---> USED TO SET A LABEL FOR THE Y-AXIS.
- plt.title("text") ---> USED TO SET A TITLE FOR THE PLOT.
- plt.legend(loc="position", fontsize=value) ---> USED TO DISPLAY A LEGEND BOX THAT IDENTIFIES WHICH PLOT REPRESENTS WHICH DATA.
- plt.grid(color="color_name", linestyle="line_style", linewidth=value) ---> USED TO DISPLAY GRID LINES ON THE PLOT.
- plt.xlim() ---> USED TO SET THE RANGE OF VALUES DISPLAYED ON THE X-AXIS.
- plt.ylim() --->USED TO SET THE RANGE OF VALUES DISPLAYED ON THE Y-AXIS.
- plt.xticks() ---> USED TO SET OR CUSTOMIZE THE VALUES DISPLAYED ON THE X-AXIS.
- plt.yticks() ---> USED TO SET OR CUSTOMIZE THE VALUES DISPLAYED ON THE Y-AXIS.
- plt.style.available ---> DISPLAYS A LIST OF ALL AVAILABLE BUILT-IN MATPLOTLIB STYLES.
- plt.style.use("style_name") ---> APPLIES A PREDEFINED STYLE TO ALL SUBSEQUENT PLOTS.
- plt.savefig('filename.extension', dpi=value, bbox_inches='tight') ---> USED TO SAVE A PLOT AS AN IMAGE FILE.
    - "filename.extension" ---> SPECIFIES THE FILE NAME AND FORMAT OF THE SAVED PLOT (e.g. .PNG, .JPG, .PDF, .SVG).
    - dpi=value ---> SETS THE RESOLUTION OF THE SAVED IMAGE IN DOTS PER INCH (HIGHER DPI = BETTER IMAGE QUALITY).
    - bbox_inches="tight" ---> AUTOMATICALLY ADJUSTS THE FIGURE BOUNDARIES TO INCLUDE ALL PLOT ELEMENTS WITHOUT EXTRA WHITESPACE.
- plt.figure(figsize=(width, height), dpi=value) ---> CREATES A NEW FIGURE (PLOTTING AREA) WITH CUSTOM SIZE AND RESOLUTION.
    - figsize=(width, height) ---> SETS THE WIDTH AND HEIGHT OF THE ENTIRE FIGURE IN INCHES.

```python
# Sample data: months and their corresponding sales values
month = [1,2,3,4]
sales = [1000, 1500, 1200, 1800]

# Creating a line plot with custom color, style, width, marker and label
plt.plot(month, sales, color="r", linestyle="--", linewidth=3, marker="o", label="2025 SALES DATA")

# Adding axis labels and title
plt.xlabel("Months")
plt.ylabel("Sales per Month")
plt.title("MONTHLY SALES DATA REPORT")

# Displaying legend and grid for better readability
plt.legend()
plt.grid()

# Saving the plot as a high resolution PNG file
plt.savefig("sales_report.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Sample data: months and their corresponding sales values
month = [1,2,3,4]
sales = [1000, 1500, 1200, 1800]

# Creating a line plot with large semi-transparent markers to demonstrate markersize and alpha
plt.plot(month, sales, color="r", linestyle="--", linewidth=3, marker="o", markersize=15, alpha=0.5, label="2025 SALES DATA")

# Adding axis labels, title, legend and grid
plt.xlabel("Months")
plt.ylabel("Sales per Month")
plt.title("MONTHLY SALES DATA REPORT")
plt.legend()
plt.grid()
plt.show()
```

```python
# Years for comparison
years = [2008, 2009, 2010, 2011, 2012]

# Oscar Winner Movies and their box office revenue (in million $)
oscar_movies = ["The Dark Knight", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]

# Non-Oscar Winner Movies and their box office revenue (in million $)
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

# Plotting both lines on the same chart for comparison
plt.plot(years, oscar_revenue, label="Oscar Winner Movies")
plt.plot(years, non_oscar_revenue, label="Non-Oscar Winner Movies")

# Adding axis labels, title and legend
plt.xlabel("Years")
plt.ylabel("Revenue (in M$)")
plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.legend()

# Applying dark background style for better visual contrast
plt.style.use('dark_background')
plt.show()
```
