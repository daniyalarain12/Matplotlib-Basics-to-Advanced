# SCATTER PLOT

A SCATTER PLOT SHOWS THE RELATIONSHIP BETWEEN TWO VARIABLES BY PLOTTING POINTS ON A 2D PLANE.
IT IS COMMONLY USED TO STUDY CORRELATION, IDENTIFY OUTLIERS, AND DETECT CLUSTERS OR PATTERNS IN DATA.

- plt.scatter(x, y, color=color_or_color_list, marker="marker_symbol", label="label_name", s=value, alpha=value, c=value_list, cmap="colormap_name") ---> USED TO CREATE A SCATTER PLOT.
  - color=color_or_color_list ---> ASSIGNS A SINGLE COLOR TO ALL POINTS OR DIFFERENT COLORS TO INDIVIDUAL POINTS USING A LIST OF COLORS.
  - s=value ---> SETS THE SIZE OF THE SCATTER POINTS.
  - c=value_list ---> ASSIGNS NUMERICAL VALUES TO POINTS FOR COLOR MAPPING.
  - cmap="colormap_name" ---> APPLIES A COLORMAP WHEN c CONTAINS NUMERICAL VALUES.
    
- plt.annotate("text", xy=(x_point,y_point), xytext=(x_text,y_text)) ---> USED TO ADD A TEXT LABEL OR NOTE TO A SPECIFIC POINT ON THE PLOT.
    - xy=(x_point, y_point) ---> SPECIFIES THE DATA POINT TO BE ANNOTATED.
    - xytext=(x_text, y_text) ---> SPECIFIES THE POSITION WHERE THE ANNOTATION TEXT WILL BE DISPLAYED.
- plt.colorbar(label="label_name") ---> USED TO DISPLAY A COLOR SCALE THAT EXPLAINS THE MAPPING BETWEEN COLORS AND DATA VALUES.

```python
# Importing Matplotlib's pyplot module
import matplotlib.pyplot as plt
```

```python
# Sample data: hours studied and corresponding exam scores
hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,65,70,75,80,85]

# Creating a basic scatter plot to show positive correlation between study time and score
plt.scatter(hours_studied, exam_scores, color='green', marker='o', label='Student Data')

# Adding axis labels, title, legend and grid
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid()
plt.show()
```

```python
# Sample data: people with their age and blood pressure values
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
# Basic scatter plot to visualize relationship between age and blood pressure
plt.scatter(age, bp, color='red')
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.show()
```

```python
# Creating a color list: green for normal BP (below 135), red for high BP (135 and above)
# Using list comprehension to assign colors based on BP value condition
colors = ["green" if x<135 else "red" for x in bp] 

# Scatter plot with conditional coloring to highlight high BP patients
plt.scatter(age, bp, color=colors)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.show()
```

```python
# Conditional color list: green for normal, red for high BP
colors = ["green" if x<135 else "red" for x in bp]

# Using BP values as point size (s=bp) creates a bubble effect
# Higher BP = larger bubble, alpha adds transparency to avoid overlap
plt.scatter(age, bp, color=colors, s=bp, alpha=0.5)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.show()
```

```python
# Using colormap (cmap) instead of fixed colors — c=bp maps BP values to colors automatically
# s=bp makes point size proportional to blood pressure value
plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

# Adding colorbar to explain the color-to-BP-value mapping
plt.colorbar(label="BP")
plt.show()
```

```python
# Sample data: people with their age and blood pressure values
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

# Annotating each data point with person's name using a loop
# xy = actual data point position, xytext = position of label text (slightly offset)
for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], bp[i]), xytext=(age[i]+1, bp[i]+1))

# Scatter plot with colormap and bubble size based on BP
plt.scatter(age, bp, s=bp, cmap="OrRd", c=bp)
plt.title("Age vs BP")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.colorbar(label="BP")

# Setting axis limits to give labels enough space without clipping
plt.xlim(20,80)
plt.ylim(100,160)
plt.show()
```

```python
# Sample data: study hours and exam scores for two different classes
hours_studied = [1,2,3]
exam_scores1 = [50,55,60]
exam_scores2 = [45,50,52]

# Plotting two scatter plots on the same chart to compare Class A and Class B performance
plt.scatter(hours_studied, exam_scores1, color='blue', label='Class A')
plt.scatter(hours_studied, exam_scores2, color='orange', label='Class B')

# Adding axis labels, title, legend and grid
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparison of 2 Classes')
plt.legend()
plt.grid()
plt.show()
```

```python
# Sample data: 5 cities with their temperature and humidity values for winter and summer
cities = ["City A", "City B", "City C", "City D", "City E"]

winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]

# Plotting winter and summer data as two separate scatter groups for seasonal comparison
plt.scatter(winter_temp, winter_humidity, label="Winter")
plt.scatter(summer_temp, summer_humidity, label="Summer")

# Adding title, axis labels and legend
plt.title("Summer vs Winter Data")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.legend()
plt.show()
```
