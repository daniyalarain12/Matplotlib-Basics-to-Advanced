# PIE CHART

A PIE CHART IS A CIRCULAR CHART DIVIDED INTO SLICES TO SHOW RELATIVE PROPORTIONS OF A WHOLE.
- EACH SLICE REPRESENTS A CATEGORY.
- THE SIZE OF EACH SLICE IS PROPORTIONAL TO ITS VALUE RELATIVE TO THE TOTAL.

plt.pie(values, labels=label_list, colors=color_list, autopct="%1.1f%%", radius=value, explode=explode_list, startangle=angle, shadow=True, wedgeprops = {"edgecolor":"color_name", "linewidth":value, "linestyle":"line_style", "alpha":value, "fill":False}) ---> USED TO CREATE A PIE CHART.

- labels=label_list ---> DISPLAYS LABELS FOR EACH PIE SLICE.
- colors=color_list ---> ASSIGNS CUSTOM COLORS TO THE PIE SLICES.
- autopct="%1.1f%%" ---> DISPLAYS THE PERCENTAGE VALUE OF EACH SLICE.
- radius=value ---> SETS THE RADIUS (SIZE) OF THE PIE CHART.
- explode=explode_list ---> SEPARATES OR HIGHLIGHTS SPECIFIC SLICES BY PULLING THEM AWAY FROM THE CENTER.
- startangle=angle ---> ROTATES THE PIE CHART TO START AT A SPECIFIED ANGLE.
- shadow=True ---> ADDS A SHADOW EFFECT TO GIVE THE CHART A MORE VISUAL 3D LOOK.
- wedgeprops={...} ---> CUSTOMIZES THE APPEARANCE OF PIE SLICES.

    - edgecolor="color_name" ---> SETS THE BORDER COLOR OF EACH PIE SLICE.
    - linewidth=value ---> SETS THE THICKNESS OF THE SLICE BORDER.
    - linestyle="line_style" ---> SETS THE STYLE OF THE BORDER LINE (SOLID, DASHED, DOTTED, ETC.).
    - fill=False ---> DISABLES FILLING OF PIE SLICES, SO ONLY THE SLICE BORDERS (OUTLINES) ARE DISPLAYED.

```python
# Importing Matplotlib's pyplot module
import matplotlib.pyplot as plt
```

```python
# Sample data: sales regions and their revenue contributions
regions = ["North", "South", "East", "West"]
revenue = [3000, 2000, 1500, 1000]

# Creating a basic pie chart with custom colors and percentage labels
plt.pie(revenue, labels=regions, colors=['gold','skyblue','lightgreen','coral'], autopct='%1.1f%%')
plt.title("Revenue Contribution By Region")
plt.show()
```

```python
# Sample data: company expense categories and their amounts
expenses = ["Salaries", "Rent", "Marketing","R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
```

```python
# Explode list: pulling out the Marketing slice (index 2) to highlight it
explodes = [0, 0, 0.2, 0, 0]

# Creating a pie chart with exploded slice and larger radius
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', radius=1.5, explode=explodes)
plt.show()
```

```python
# Rotating the pie chart 90 degrees so the first slice starts from the top
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', startangle=90)
plt.show()
```

```python
# Adding a shadow effect to give the pie chart a 3D appearance
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
# Customizing pie slice borders using wedgeprops: black dashed border with thickness 2
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', wedgeprops={
    "edgecolor" : "black",
    "linestyle" : "--",
    "linewidth" : 2
})
plt.show()
```

```python
# Making pie slices semi-transparent using alpha (0.5 = 50% transparency)
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', wedgeprops={
    "alpha" : 0.5
})
plt.show()
```

```python
# Disabling fill to show only the outlines of pie slices (hollow pie chart)
plt.pie(amounts, labels=expenses, autopct='%1.1f%%', wedgeprops={
    "fill" : False
})
plt.show()
```
