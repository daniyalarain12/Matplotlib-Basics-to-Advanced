# HISTOGRAMS

A histogram shows the distribution of numerical data by:
- Dividing values into bins (ranges)
- Counting how many values fall into each bin (frequency)

plt.hist(data, bins=Number_of_bins,  color="color_name",  edgecolor="color_name", label="label_name", alpha=value) ---> USED TO CREATA A HISTOGRAM.
- bins=Number_of_bins ---> DIVIDES THE DATA INTO A SPECIFIED NUMBER OF INTERVALS (BINS).
- bins=[bin_edges] ---> DEFINES CUSTOM BIN RANGES BY SPECIFYING THE EXACT BIN BOUNDARIES.

plt.axvline(x=value, color="color_name", linestyle="line_style", linewidth=value, label="label_name") ---> USED TO DRAW A VERTICAL REFERENCE LINE ON A PLOT.
- x=value ---> SPECIFIES THE X-AXIS POSITION WHERE THE VERTICAL LINE WILL BE DRAWN.

```python
# Importing Matplotlib's pyplot module
import matplotlib.pyplot as plt
```

```python
# Sample data: exam scores of 20 students
scores = [45,67,89,56,78,88,92,60,74,81,59,66,75,82,90,85,70,73,68,77]

# Creating a histogram with 5 bins to show score distribution
# edgecolor adds black borders to each bin for better visibility
plt.hist(scores, bins=5, color='purple', edgecolor='black')

# Adding axis labels and title
plt.xlabel("Score Range")
plt.ylabel("Number of Students")
plt.title("Score Distribution of Students")
plt.show()
```

```python
# Sample data: marks of 30 students
marks = [25,38,42,55,61,67,72,78,83,88,91,95,47,53,59,64,69,74,79,85,92,97,33,45,58,62,71,76,81,89]

# Defining custom bin edges: 0-30 (fail), 30-70 (average), 70-90 (good), 90-100 (excellent)
bins = [0, 30, 70, 90, 100]

# Adding axis labels and title
plt.xlabel("Marks")
plt.ylabel("No. of Students")
plt.title("Distribution of Marks for Students")

# Creating histogram with custom bin ranges instead of equal-width bins
plt.hist(marks, color="skyblue", edgecolor="black", bins=bins)
plt.show()
```

```python
# Sample data: transaction amounts for legitimate and fraudulent transactions
legit_transactions = [2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50, 45.00, 49.99, 55.25, 
                      60.00, 75.99, 89.99, 120.50, 150.00, 199.99, 249.99, 300.75, 450.00, 600.00, 850.00, 1200.00]
fraud_transactions = [50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000, 1200, 1500, 1500, 2000, 2500, 3000, 3000]

# Overlapping both histograms using alpha for transparency so both are visible
plt.hist(legit_transactions, color="green", edgecolor="black", bins=10, label="Legit", alpha=0.5)
plt.hist(fraud_transactions, color="red", edgecolor="black", bins=10, label="Fraud", alpha=0.5)

# Adding axis labels, title and legend to differentiate both distributions
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.title("Transaction Amount Distribution - Legit vs Fraud")
plt.legend()
plt.show()
```

```python
# Sample data: marks of 30 students
marks = [25,38,42,55,61,67,72,78,83,88,91,95,47,53,59,64,69,74,79,85,92,97,33,45,58,62,71,76,81,89]

# Custom bin ranges for grade categories
bins = [0, 30, 70, 90, 100]

# Adding axis labels and title
plt.xlabel("Marks")
plt.ylabel("No. of Students")
plt.title("Distribution of Marks for Students")

# Creating histogram with custom bins
plt.hist(marks, color="skyblue", edgecolor="black", bins=bins)

# Adding a vertical reference line at passing marks (33) for quick visual reference
plt.axvline(x=33, color='red', linestyle='--', linewidth=2, label="Passing Marks (33)")

plt.legend()
plt.show()
```
