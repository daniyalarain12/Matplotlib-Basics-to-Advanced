# BOX PLOT

A BOX PLOT IS A STATISTICAL VISUALIZATION THAT SUMMARIZES THE DISTRIBUTION OF A DATASET USING FIVE KEY NUMBERS:
- MINIMUM (LOWEST NON-OUTLIER) ---> LOWER WHISKER
- FIRST QUARTILE (Q1) ---> 25TH PERCENTILE
- MEDIAN (Q2) ---> 50TH PERCENTILE
- THIRD QUARTILE (Q3) ---> 75TH PERCENTILE
- MAXIMUM (HIGHEST NON-OUTLIER) ---> UPPER WHISKER
    - LOWER WHISKER = Q1 - (1.5 * IQR)
    - UPPER WHISKER = Q3 + (1.5 * IQR)
        - IQR = Q3 - Q1 ---> INTER-QUARTILE RANGE
        - 1.5 ---> DEFAULT WHISKER LENGTH MULTIPLIER

COMMONLY USED TO IDENTIFY OUTLIERS, ANALYZE DATA SPREAD, DETECT SKEWNESS, AND COMPARE DISTRIBUTIONS ACROSS MULTIPLE DATASETS.

- plt.boxplot(data_or_data_list, labels=label_list, showmeans=True, vert=False, whis=value) ---> USED TO CREATE A BOX PLOT.

    - data_or_data_list ---> SPECIFIES A SINGLE DATASET OR A LIST OF DATASETS TO BE DISPLAYED AS BOX PLOTS.
    - labels=label_list ---> ASSIGNS LABELS TO THE BOX PLOTS.
    - showmeans=True ---> DISPLAYS THE MEAN (AVERAGE) VALUE ON THE BOX PLOT.
    - vert=False ---> DISPLAYS THE BOX PLOT HORIZONTALLY. IF True, THE BOX PLOT IS DISPLAYED VERTICALLY.
    - whis=value ---> CONTROLS THE LENGTH OF THE WHISKERS AS A MULTIPLE OF THE IQR.

```python
# Importing Matplotlib's pyplot module
import matplotlib.pyplot as plt
```

```python
# Sample data: a single dataset to demonstrate basic box plot structure
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

# Creating a basic vertical box plot showing Q1, median, Q3, whiskers
plt.boxplot(data)
plt.grid()
plt.show()
```

```python
# Same dataset as above
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

# showmeans=True adds a marker for the mean value inside the box
# vert=False rotates the box plot to display horizontally
plt.boxplot(data, showmeans=True, vert=False)
plt.grid()
plt.show()
```

```python
# Same dataset as above
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

# whis=1 reduces whisker length to 1x IQR instead of default 1.5x
# This makes more points appear as outliers compared to default
plt.boxplot(data, whis=1)
plt.grid()
plt.show()
```

```python
# Modified dataset: value 15 replaced with 20 to introduce a clear outlier
data = [7, 8, 5, 6, 9, 4, 10, 12, 20]

# Box plot showing 20 as an outlier (plotted as a circle beyond the whisker)
plt.boxplot(data)
plt.grid()
plt.show()
```

```python
# Two separate datasets to compare their distributions side by side
group1 = [7, 8, 5, 6, 9, 4, 10, 12, 20]
group2 = [4, 9, 13, 15, 3, 30, 15, 12]

# Passing both datasets as a list to display two box plots in one figure
plt.boxplot([group1, group2])
plt.grid()
plt.show()
```

```python
# Two datasets for group comparison
group1 = [7, 8, 5, 6, 9, 4, 10, 12, 20]
group2 = [4, 9, 13, 15, 3, 30, 15, 12]

# Adding labels to identify each box plot and showmeans to display mean markers
plt.boxplot([group1, group2], labels=["Group1","Group2"], showmeans=True)
plt.grid()
plt.show()
```
