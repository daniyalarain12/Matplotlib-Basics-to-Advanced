# STACK PLOT

- A stack plot is a type of plot where multiple data series are stacked on top of each other.
- Each layer shows the contribution of one category, and the top line shows the cumulative total.

plt.stackplot(x_values, y1, y2, y3, ..., labels=label_list, colors=color_list, alpha=value)

- x_values ---> SPECIFIES THE VALUES FOR THE X-AXIS.
- y1, y2, y3, ... ---> SPECIFY THE DATA SERIES TO BE STACKED.
- labels=label_list ---> ASSIGNS LABELS TO EACH STACKED DATA SERIES.

```python
# Importing Matplotlib's pyplot module
import matplotlib.pyplot as plt
```

```python
# Sample data: days of the week and customer counts from 3 different marketing channels
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
direct = [50, 60, 70, 80, 90, 100, 110]    # Customers from direct visits
organic = [30, 40, 50, 55, 60, 70, 80]     # Customers from organic search
social = [20, 25, 30, 35, 40, 50, 60]      # Customers from social media
```

```python
# Creating a stack plot to show combined and individual contribution of each marketing channel
# Each layer stacks on top of the previous one - top of chart shows total customers per day
plt.stackplot(days, direct, organic, social, labels=['Direct','Organic','Social'])

# Adding axis labels, title, legend and grid
plt.xlabel('Days')
plt.ylabel('No. of Customers')
plt.title('Marketing Data for the Week')
plt.legend()
plt.grid()
plt.show()
```
