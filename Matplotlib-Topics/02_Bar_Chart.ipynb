# BAR CHART

A bar plot (or bar chart) is a graph that displays categorical data using rectangular bars.
Each bar's height (or length, in horizontal bars) represents a value.

- plt.bar(x, height, color=color_or_color_list, width=value, label="label_name", alpha=value) ---> USED TO CREATE A VERTICAL BAR CHART.
- plt.barh(y, width, color=color_or_color_list, height=value, label="label_name", alpha=value) ---> USED TO CREATE A HORIZONTAL BAR CHART.
- plt.text(x_position, y_position, "text", ha="alignment") ---> USED TO ADD CUSTOM TEXT AT A SPECIFIED POSITION ON THE PLOT WITH OPTIONAL TEXT ALIGNMENT.

```python
# Importing required libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data: product names and their sales values
product = ["A","B","C","D"]
sales = [1000, 1500, 800, 1200]

# Creating a vertical bar chart with a single green color for all bars
plt.bar(product, sales, color="g", label="SALES 2026")

# Adding axis labels, title and legend
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("PRODUCT SALE COMPARISON")
plt.legend()
plt.show()
```

```python
# Sample data: product names and their sales values
product = ["A","B","C","D"]
sales = [1000, 1500, 800, 1200]

# Creating a horizontal bar chart with different colors for each bar
plt.barh(product, sales, color=['yellow','green','blue','red'], label="SALES 2026")

# Adding axis labels, title and legend
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("PRODUCT SALE COMPARISON")
plt.legend()
plt.show()
```

```python
# Oscar Winner Movies data
movies = ["The Dark Knight", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]

# Creating a vertical bar chart for Oscar movie revenues
plt.bar(years, oscar_revenue, color="r")
plt.xlabel("Product")
plt.ylabel("Revenue (in $M)")
plt.title("REVENUE IN EACH YEAR FOR OSCAR MOVIES")

# Adding revenue value labels on top of each bar using a loop
# x position = year, y position = revenue + 20 (slight offset above bar)
for i in range(len(years)):
    plt.text(years[i], oscar_revenue[i] + 20, str(oscar_revenue[i]), ha="center")

# Setting y-axis limit slightly above max value to accommodate text labels
plt.ylim(0,1100)
plt.show()
```

```python
# Years for comparison
years = [2008, 2009, 2010, 2011, 2012]

# Oscar Winner Movies and their box office revenue (in million $)
oscar_movies = ["The Dark Knight", "The Hurt Locker", "The King's Speech", "The Artist", "Argo"]
oscar_revenue = [1005, 170, 427, 133, 232]

# Non-Oscar Winner Movies and their box office revenue (in million $)
non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]

# Creating evenly spaced positions for grouped bars using NumPy
x = np.arange(len(oscar_movies))
width = 0.4  # Width of each bar

# Plotting two bars side by side: shifting left and right by half the width
plt.bar(x - width/2, oscar_revenue, width, label="Oscar Winner Movies")
plt.bar(x + width/2, non_oscar_revenue, width, label="Non-Oscar Winner Movies")

# Adding axis labels, title and legend
plt.xlabel("Years")
plt.ylabel("Revenue (in M$)")
plt.title("Oscar Movies vs Non-Oscar Movies Revenue")
plt.legend()

# Replacing numeric x-axis positions with actual year labels
plt.xticks(x, years)
plt.show()
```
