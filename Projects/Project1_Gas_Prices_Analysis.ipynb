# Gas Prices Analysis — Tracking Fuel Costs Across Countries Over the Years

### Objective
Analyze and visualize how gas prices have changed over time across multiple countries using line plots.

### Dataset
- **File:** gas_prices.csv
- **Rows:** 19 (years from 1990 to 2008)
- **Columns:** 11 (Year + 10 countries)
- **Countries Covered:** Australia, Canada, France, Germany, Italy, Japan, Mexico, South Korea, UK, USA
- **Values:** Annual gas prices in USD

### Tools Used
- **Pandas** — Loading and accessing dataset
- **Matplotlib** — Data visualization

```python
# Importing required libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Loading the gas prices dataset into a DataFrame
# Dataset contains annual gas prices (in USD) for 10 countries from 1990 to 2008
df = pd.read_csv('gas_prices.csv')
```

## Analysis 1 — USA Gas Price Trend (1990-2008)

```python
# Plotting USA gas prices over all years (1990-2008)
plt.plot(df['Year'], df['USA'], label='USA', marker='o')

# Adding axis labels and title
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('USA Gas Prices (1990-2008) in USD')
plt.legend()

# Showing every 3rd year on x-axis to avoid overcrowding of tick labels
plt.xticks(df.Year[::3])
plt.show()
```

## Analysis 2 — USA vs Canada vs South Korea Gas Price Comparison (1990-2008)

```python
# Plotting 3 countries on the same chart for a direct side-by-side comparison
# Each country gets a distinct color for easy visual differentiation
plt.plot(df['Year'], df['USA'], label='USA', color='r', marker='o')
plt.plot(df['Year'], df['Canada'], label='Canada', color='b', marker='o')
plt.plot(df['Year'], df['South Korea'], label='South Korea', color='g', marker='o')

# Adding axis labels and title
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('USA vs Canada vs South Korea — Gas Prices (1990-2008) in USD')
plt.legend()

# Showing every 3rd year on x-axis to avoid overcrowding
plt.xticks(df.Year[::3])
plt.show()
```

## Analysis 3 — Gas Price Trends for All 10 Countries (1990-2008)

```python
# Larger figure size to accommodate all 10 country lines and legend clearly
plt.figure(figsize=(10,8))

# Dynamically plotting all 10 countries using a loop
# Skipping the 'Year' column since it is the x-axis, not a country
for country in df.columns:
    if (country != 'Year'):
        plt.plot(df['Year'], df[country], label=country, marker='o')

# Adding axis labels, title and legend
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('Global Gas Prices — All 10 Countries (1990-2008) in USD')
plt.legend()

# Showing every 3rd year to prevent x-axis label clutter
plt.xticks(df.Year[::3])
plt.show()
```

## Analysis 4 — Gas Price Comparison for 5 Major Countries (1990-2008)

```python
plt.figure(figsize=(8,5))

# Selecting 5 major countries from different continents for a focused comparison
# Australia (Oceania), USA (North America), Canada (North America), UK (Europe), Japan (Asia)
countries = ['Australia', 'USA', 'Canada', 'UK', 'Japan']

# Plotting each selected country dynamically using a loop
for country in countries:
    plt.plot(df['Year'], df[country], label=country, marker='o')

# Adding axis labels, title and legend
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.title('Gas Price Comparison — Australia, USA, Canada, UK & Japan (1990-2008) in USD')
plt.legend()

# Showing every 3rd year on x-axis
plt.xticks(df.Year[::3])
plt.show()
```
